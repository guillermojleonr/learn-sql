```python
import sqlite3

#conn
conn = sqlite3.connect('example.db')
```

```python
query = """
CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT, age INTEGER)
"""
conn.execute(query)
```

```python
query = """
INSERT INTO users (name, age) VALUES ('Alice', 30), ('Bob', 25), ('Charlie', 35)
"""
conn.execute(query)
conn.commit()
```

```python
query = """
SELECT * FROM users
"""
cursor = conn.execute(query)
for row in cursor:
    print(row)
```
